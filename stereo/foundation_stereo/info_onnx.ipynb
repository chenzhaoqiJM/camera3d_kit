```python
import onnxruntime as ort

def print_onnx_io(model_path):
    # 创建推理 Session
    session = ort.InferenceSession(model_path, providers=["CPUExecutionProvider"])

    # 打印输入信息
    print("=== Model Inputs ===")
    for idx, inp in enumerate(session.get_inputs()):
        print(f"[{idx}] name : {inp.name}")
        print(f"    type : {inp.type}")
        print(f"    shape: {inp.shape}")

    # 打印输出信息
    print("\n=== Model Outputs ===")
    for idx, out in enumerate(session.get_outputs()):
        print(f"[{idx}] name : {out.name}")
        print(f"    type : {out.type}")
        print(f"    shape: {out.shape}")

# 使用示例
# print_onnx_io(r"C:\Users\86994\Desktop\foundation_stereo_23-51-11.onnx")

print_onnx_io("./foundation_stereo_small.onnx")

```
